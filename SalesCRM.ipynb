```python
import pandas as pd
```

```python
account = pd.read_excel("HireArt_Data_Analyst_Exercise.xlsx")
```

```python
account.head()
```

```python
account
```

```python
account['month'] = pd.DatetimeIndex(account["Date of Contact"]).month
```

```python
account.head()
```

```python
account['Month_Year'] = account['Date of Contact'].dt.strftime('%b-%Y')
```

```python
d = {1 : 'January', 2 : 'February', 3 : 'March', 4: 'April', 5: 'May', 6: 'June', 7 : 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
```

```python
account['Month_Name'] = account['month'].map(d)
```

```python
Month_count_df = account.groupby("Month_Name").count()
```

```python
Month_count_df
```

```python
%matplotlib inline
```

```python
Month_count_df.plot(kind='bar')
```

```python
Month_count_df.plot(y = "Account manager", kind='bar', figsize=(16, 9)).legend(bbox_to_anchor=(0.2, 0.9))
#plot.bar.figsize=(16, 9)
```

```python
Month_count_df["Percentage of its clients"] = Month_count_df["Client Name"] / Month_count_df["Client Name"].sum()
```

```python
Month_count_df
```

```python
Month_count_df["Percentage of its clients"].sum()
```

```python
121 / 1001
```

```python
Month_count_df.sort_values('Percentage of its clients', ascending=False)[['Percentage of its clients']].plot(y = "Percentage of its clients", kind='bar', figsize=(16, 9))
```

```python
Month_year_df = account.groupby("Month_Year").count()
```

```python
Month_year_df
```

```python
Month_year_df.sort_values('Client Name', ascending=False)[['Client Name']].plot(y = "Client Name", kind='bar', figsize=(16, 9))
```
